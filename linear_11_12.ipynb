```python
# Data
heights = [147, 150, 153, 155, 158, 160, 163, 165, 168, 170, 173, 175, 178, 180, 183]
weights = [49, 50, 51, 52, 54, 56, 58, 59, 60, 72, 63, 64, 66, 67, 68]
```

```python
# Function to calculate linear regression
def calculate_linear_regression(x, y):
  n = len(x)
  sum_x = sum(x)
  sum_y = sum(y)
  sum_xy = sum(xi * yi for xi, yi in zip(x, y))
  sum_xx = sum(xi * xi for xi in x)

  #Calculate m and b
  m = (n * sum_xy - sum_x * sum_y) / (n * sum_xx - sum_x ** 2)
  b = (sum_y - m * sum_x) / n
  return m, b
```

```python
# Function to predict weight using linear regression model
def predict_weight(height, m, b):
  return m * height + b
```

```python
# Main function to perform regression and predictions
def linear_regression_analysis(heights, weights):
  # Calculate regression coefficients
  m, b = calculate_linear_regression(heights, weights)
  print(f"Linear Regression equation: Weight = {m:.4f} * Height + {b:.2f}")

  # Get user inputs for prediction
  height_input = float(input("Enter the height to predict weight: "))
  predicted_weight = predict_weight(height_input, m, b)
  print(f"Predicted weight for height: {height_input} cm, {predicted_weight:.2f} kg")
```

```python
# Run predicted heights and weights based on Linear regression
linear_regression_analysis(heights, weights)
```
